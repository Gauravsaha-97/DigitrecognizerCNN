```
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from keras.utils import to_categorical
```

```
test_data = pd.read_csv("/content/drive/My Drive/Digit Recognizer/test.csv")
train_data  = pd.read_csv("/content/drive/My Drive/Digit Recognizer/train.csv")
train_data.head()
```

```
train_images = train_data.drop(columns='label', axis=1)
train_label = train_data[['label']]
train_images = train_images/255
train_images.head()
```

```
test_data.head()
```

```
test_data = test_data/255
test_data.head()
```

```
train_label.head()
```

```
train_label = to_categorical(train_label, num_classes=10)
```

```
train_label.shape
```

```
train_images = train_images.values.reshape((-1,28,28,1))
```

```
train_images.shape
```

```
test_images = test_data.values.reshape((-1,28,28,1))
test_images.shape
```

```
model = tf.keras.models.Sequential([
                                    tf.keras.layers.Conv2D(64,(3,3),activation='relu',input_shape=(28,28,1)),
                                    tf.keras.layers.MaxPooling2D(2,2),
                                    tf.keras.layers.Conv2D(128, (3,3),activation='relu'),
                                    tf.keras.layers.MaxPooling2D(2,2),
                                    tf.keras.layers.Flatten(),
                                    tf.keras.layers.Dense(512,activation='relu'),
                                    tf.keras.layers.Dense(10,activation='softmax')
])
```

```
model.compile(optimizer='rmsprop', loss='categorical_crossentropy',metrics=['accuracy'])
```

```
model.fit(train_images,train_label,batch_size=86, verbose=2, epochs=10)
```

```
pred_labels = model.predict(test_images, batch_size=86)
```

```
sample_sub = pd.read_csv('/content/drive/My Drive/Digit Recognizer/sample_submission.csv')
sample_sub.head()
```

```
pred_labels = pd.DataFrame(pred_labels)
pred_labels.head()
```

```
pred_class = model.predict_classes(test_images)
pred_class
```

```
pred_class = pd.DataFrame(pred_class)
pred_class.head()
```

```
submission_2 = pd.concat([sample_sub['ImageId'],pred_class],axis=1)
submission_2.head()
submission_2.columns=['ImageId','Label']
```
